## ðŸ§  Ejercicio 2

Un pangrama es una oraciÃ³n que utiliza todas las letras del alfabeto.

Escribe un programa que verifique si una cadena dada es un pangrama o no,
utilizando una funciÃ³n definida por el usuario `ispangram()`.


````python
# SoluciÃ³n ejercicio 2
import string

def ispangram(texto: str) -> bool:
    """Devuelve True si 'texto' es un pangrama (contiene a-z al menos una vez)."""
    texto = texto.lower()
    letras_en_texto = {c for c in texto if c.isalpha()}
    return set(string.ascii_lowercase).issubset(letras_en_texto)

# Ejemplos de prueba
print(ispangram("The quick brown fox jumps over a lazy dog"))  # True
print(ispangram("Hola mundo"))  # False
````

## ðŸ§  Ejercicio 3

Escribe un programa en Python que acepte una secuencia de palabras separadas
por guiones como entrada y llame a una funciÃ³n `convert()` que la convierta en
una secuencia separada por guiones despuÃ©s de ordenar las palabras alfabÃ©ticamente.

Ejemplo:

Entrada:
`here-come-the-dots-followed-by-dashes`

Salida:
`by-come-dashes-dots-followed-here-the`


````python
# SoluciÃ³n ejercicio 3

def convert(cadena: str) -> str:
    """Ordena alfabÃ©ticamente las palabras separadas por '-' y las vuelve a unir."""
    partes = cadena.split("-")
    partes_ordenadas = sorted(partes)
    return "-".join(partes_ordenadas)

entrada = "here-come-the-dots-followed-by-dashes"
salida = convert(entrada)
print(salida)
````

## ðŸ§  Ejercicio 5

Un palÃ­ndromo es una palabra o frase que se lee igual en ambas direcciones.
Ejemplos: `deed`, `level`, `Malayalam`, `Rats live on no evil star`.

Escribe una funciÃ³n `ispalindrome()` que verifique si una cadena dada es un
palÃ­ndromo o no. Ignora espacios y diferencias entre mayÃºsculas y minÃºsculas.


````python
# SoluciÃ³n ejercicio 5

def ispalindrome(texto: str) -> bool:
    """Devuelve True si 'texto' es un palÃ­ndromo, ignorando espacios y mayÃºsculas."""
    # Dejamos solo caracteres alfanumÃ©ricos
    filtrado = "".join(c.lower() for c in texto if c.isalnum())
    return filtrado == filtrado[::-1]

# Pruebas
print(ispalindrome("deed"))
print(ispalindrome("Rats live on no evil star"))
print(ispalindrome("Hola mundo"))
````

## ðŸ§  Ejercicio 6

Escribe una funciÃ³n `convert()` que reciba una cadena con palabras separadas
por espacios y devuelva una cadena despuÃ©s de eliminar todas las palabras
duplicadas y ordenarlas alfanumÃ©ricamente.

Ejemplo de cadena:
```python
s = "Sakhi was a singer because her mother was a singer, and Sakhi's mother was a singer because her father was a singer"
```


````python
# SoluciÃ³n ejercicio 6
import string as _string

def convert_unique_sorted(cadena: str) -> str:
    """Elimina palabras duplicadas, limpia puntuaciÃ³n simple y ordena alfabÃ©ticamente."""
    palabras_crudas = cadena.split()
    # Limpiamos puntuaciÃ³n bÃ¡sica al final o inicio de palabra
    palabras_limpias = [
        palabra.strip(_string.punctuation).lower()
        for palabra in palabras_crudas
        if palabra.strip(_string.punctuation)
    ]
    unicas = sorted(set(palabras_limpias))
    return " ".join(unicas)

s = "Sakhi was a singer because her mother was a singer, and Sakhi's mother was a singer because her father was a singer"
print(convert_unique_sorted(s))
````

````python
# SoluciÃ³n ejercicio 7

def count_alphabets_digits(cadena: str) -> dict:
    """Cuenta letras y dÃ­gitos en una cadena y devuelve un diccionario."""
    letras = sum(1 for c in cadena if c.isalpha())
    digitos = sum(1 for c in cadena if c.isdigit())
    return {"letters": letras, "digits": digitos}

print(count_alphabets_digits("Python 3.11 es genial, versiÃ³n 2025"))
````

## ðŸ§  Ejercicio 8

Escribe una funciÃ³n `frequency()` que calcule la frecuencia de las palabras
presentes en una cadena. Las frecuencias deben devolverse ordenadas
alfabÃ©ticamente por palabra.


````python
# SoluciÃ³n ejercicio 8
import string as _str

def frequency(texto: str) -> dict:
    """Devuelve un diccionario palabra -> frecuencia, ordenado alfabÃ©ticamente."""
    # Limpiamos puntuaciÃ³n simple y pasamos a minÃºsculas
    trad = str.maketrans("", "", _str.punctuation)
    limpio = texto.lower().translate(trad)
    palabras = limpio.split()
    frec = {pal: palabras.count(pal) for pal in set(palabras)}
    # Reconstruimos diccionario ordenado por palabra
    frec_ordenado = {pal: frec[pal] for pal in sorted(frec.keys())}
    return frec_ordenado

texto_ej = "Python es genial y Python es poderoso"
print(frequency(texto_ej))
````

## ðŸ§  Ejercicio 9

Escribe dos funciones `create_sent1()` y `create_sent2()`.
Ambas reciben 3 listas:

```python
subjects = ['He', 'She']
verbs = ['loves', 'hates']
objects = ['TV Serials', 'Netflix']
```

Ambas funciones deben formar oraciones seleccionando elementos de estas listas
y devolverlas (todas las combinaciones posibles).


````python
# SoluciÃ³n ejercicio 9
subjects = ['He', 'She']
verbs = ['loves', 'hates']
objects = ['TV Serials', 'Netflix']

def create_sent1(subj, verbs, objs):
    """Genera todas las combinaciones sujeto-verbo-objeto con bucles normales."""
    oraciones = []
    for s in subj:
        for v in verbs:
            for o in objs:
                oraciones.append(f"{s} {v} {o}")
    return oraciones

def create_sent2(subj, verbs, objs):
    """Genera todas las combinaciones usando list comprehension."""
    return [f"{s} {v} {o}" for s in subj for v in verbs for o in objs]

print(create_sent1(subjects, verbs, objects))
print(create_sent2(subjects, verbs, objects))
````

## ðŸ§  Ejercicio 10

Escriba un programa que le pida al usuario introducir valores (x, y, z) y,
segÃºn estos valores, imprima si pertenece al plano (z = 0) o al espacio (z â‰  0),
indicando tambiÃ©n a quÃ© cuadrante (en el plano) u octante (en el espacio) pertenece.


````python
# SoluciÃ³n ejercicio 10

def clasificar_punto(x: float, y: float, z: float) -> str:
    """Clasifica un punto en el plano (z=0) o en el espacio (z!=0),
    indicando cuadrante u octante.
    """
    if z == 0:
        # Estamos en el plano XY
        if x == 0 or y == 0:
            return "El punto estÃ¡ en el plano (z=0) pero sobre un eje, no en un cuadrante."
        if x > 0 and y > 0:
            cuadrante = 1
        elif x < 0 and y > 0:
            cuadrante = 2
        elif x < 0 and y < 0:
            cuadrante = 3
        else:
            cuadrante = 4
        return f"El punto estÃ¡ en el plano y pertenece al cuadrante {cuadrante}."
    else:
        # Estamos en el espacio 3D
        # Octante segÃºn el signo de (x, y, z)
        signos = (x >= 0, y >= 0, z >= 0)
        return f"El punto estÃ¡ en el espacio 3D con signos (x>=0, y>=0, z>=0) = {signos}."

# Ejemplo de uso
print(clasificar_punto(1, 2, 0))
print(clasificar_punto(-1, 2, 0))
print(clasificar_punto(1, -2, 3))
````

````python
# SoluciÃ³n ejercicio 11 - Funciones de procesamiento de texto
import re
import string as _st
from collections import Counter


def contar_palabras(texto: str) -> int:
    """Cuenta el nÃºmero de palabras separadas por espacios."""
    return len(texto.split())


def contar_caracteres(texto: str, incluir_espacios: bool = True) -> int:
    """Cuenta caracteres; si incluir_espacios=False, solo cuenta letras y dÃ­gitos."""
    if incluir_espacios:
        return len(texto)
    return sum(1 for c in texto if c.isalnum())


def limpiar_texto(texto: str) -> str:
    """Convierte a minÃºsculas, elimina puntuaciÃ³n bÃ¡sica, nÃºmeros y espacios extra."""
    texto = texto.lower()
    # Quitar signos de puntuaciÃ³n bÃ¡sicos y nÃºmeros
    texto = re.sub(r"[0-9]", "", texto)
    texto = texto.translate(str.maketrans("", "", ".,!?;:"))
    # Reemplazar mÃºltiples espacios por uno
    texto = re.sub(r"\s+", " ", texto)
    return texto.strip()


def palabra_mas_frecuente(texto: str) -> str | None:
    """Devuelve la palabra mÃ¡s frecuente despuÃ©s de limpiar el texto.
    Si no hay palabras, devuelve None.
    """
    limpio = limpiar_texto(texto)
    if not limpio:
        return None
    palabras = limpio.split()
    cont = Counter(palabras)
    return cont.most_common(1)[0][0]


def contar_mayusculas(texto: str) -> int:
    """Cuenta letras mayÃºsculas en el texto original."""
    return sum(1 for c in texto if c.isupper())


_palabras_positivas = {
    "excelente", "genial", "encanta", "recomiendo", "mejor", "bueno",
    "increÃ­ble", "fascinante", "alucinado", "recomendado", "increible"
}


def es_positivo(texto: str) -> bool:
    """Devuelve True si el texto contiene alguna palabra positiva definida."""
    limpio = limpiar_texto(texto)
    palabras = set(limpio.split())
    return any(p in palabras for p in _palabras_positivas)


def extraer_hashtags(texto: str) -> list:
    """Devuelve una lista con todas las palabras que empiezan por '#'"""
    palabras = texto.split()
    return [p for p in palabras if p.startswith("#")]


def porcentaje_mayusculas(texto: str) -> float:
    """Calcula el porcentaje de letras mayÃºsculas sobre el total de letras."""
    letras = [c for c in texto if c.isalpha()]
    if not letras:
        return 0.0
    mayus = [c for c in letras if c.isupper()]
    return (len(mayus) / len(letras)) * 100

# Ejemplos rÃ¡pidos
print(contar_palabras(comentario_1))
print(limpiar_texto(comentario_1))
print(palabra_mas_frecuente(comentario_1))
print(contar_mayusculas(comentario_2))
print(es_positivo(comentario_3))
print(es_positivo(comentario_4))
print(extraer_hashtags(comentario_4))
print(porcentaje_mayusculas(tweet_curso))
````

## Ejercicio 12 - ValidaciÃ³n de Datos de Sensores

Se trabaja con una lista de temperaturas con errores (-999). Se deben crear
funciones para limpiar errores, convertir a Fahrenheit, calcular promedio y
contar errores.


````python
# SoluciÃ³n ejercicio 12

temperaturas_celsius = [
    22.5,  # 00:00
    21.8,  # 01:00
    -999,  # 02:00 - Error de sensor
    19.5,  # 03:00
    18.2,  # 04:00
    -999,  # 05:00 - Error de sensor
    17.8,  # 06:00
    19.3,  # 07:00
    21.5,  # 08:00
    24.7,  # 09:00
    -999,  # 10:00 - Error de sensor
    28.3,  # 11:00
    30.5,  # 12:00
    -999,  # 13:00 - Error de sensor
    31.2,  # 14:00
    30.8,  # 15:00
    -999,  # 16:00 - Error de sensor
    29.4,  # 17:00
    27.6,  # 18:00
    25.3,  # 19:00
    23.8,  # 20:00
    -999,  # 21:00 - Error de sensor
    22.1,  # 22:00
    21.5   # 23:00
]


def limpiar_errores(lista_temperaturas: list[float]) -> list[float]:
    """Devuelve una nueva lista sin los valores -999."""
    return [t for t in lista_temperaturas if t != -999]


def celsius_a_fahrenheit(celsius: float) -> float:
    """Convierte temperatura de Celsius a Fahrenheit."""
    return (celsius * 9 / 5) + 32


def temperatura_promedio(lista_temperaturas: list[float]) -> float:
    """Calcula el promedio de las temperaturas vÃ¡lidas (ignorando errores)."""
    limpias = limpiar_errores(lista_temperaturas)
    if not limpias:
        return 0.0
    return sum(limpias) / len(limpias)


def contar_errores(lista_temperaturas: list[float]) -> int:
    """Cuenta cuÃ¡ntos valores de error (-999) hay en la lista."""
    return sum(1 for t in lista_temperaturas if t == -999)

# Ejemplos
print("Lecturas vÃ¡lidas:", limpiar_errores(temperaturas_celsius))
print("Promedio Celsius:", temperatura_promedio(temperaturas_celsius))
print("Errores:", contar_errores(temperaturas_celsius))
print("Una conversiÃ³n de ejemplo (25C):", celsius_a_fahrenheit(25))
````

````python
# SoluciÃ³n ejercicio 13 - Funciones de anÃ¡lisis de ventas

def calcular_ingresos_totales(lista_ventas: list[dict]) -> float:
    """Suma precio * cantidad de cada producto."""
    return sum(v["precio"] * v["cantidad"] for v in lista_ventas)


def producto_mas_vendido(lista_ventas: list[dict]) -> str:
    """Devuelve el nombre del producto con mayor cantidad vendida."""
    if not lista_ventas:
        return ""
    mas = max(lista_ventas, key=lambda v: v["cantidad"])
    return mas["producto"]


def calcular_ticket_promedio(lista_ventas: list[dict]) -> float:
    """Ingreso promedio por producto (lÃ­nea de venta)."""
    if not lista_ventas:
        return 0.0
    total = calcular_ingresos_totales(lista_ventas)
    return total / len(lista_ventas)


def ventas_por_categoria(lista_ventas: list[dict]) -> dict:
    """Devuelve un diccionario categoria -> ingresos totales."""
    resultado: dict[str, float] = {}
    for v in lista_ventas:
        cat = v["categoria"]
        ingreso = v["precio"] * v["cantidad"]
        resultado[cat] = resultado.get(cat, 0) + ingreso
    return resultado


def categoria_top(lista_ventas: list[dict]) -> str:
    """Devuelve la categorÃ­a con mayores ingresos totales."""
    vp = ventas_por_categoria(lista_ventas)
    if not vp:
        return ""
    return max(vp.items(), key=lambda x: x[1])[0]


def productos_premium(lista_ventas: list[dict], precio_minimo: float = 100) -> list[str]:
    """Lista de nombres de productos con precio >= precio_minimo."""
    return [v["producto"] for v in lista_ventas if v["precio"] >= precio_minimo]


def calcular_porcentaje_categoria(lista_ventas: list[dict], categoria: str) -> float:
    """Calcula quÃ© porcentaje de los ingresos totales corresponde a una categorÃ­a dada."""
    total = calcular_ingresos_totales(lista_ventas)
    if total == 0:
        return 0.0
    vp = ventas_por_categoria(lista_ventas)
    ingreso_cat = vp.get(categoria, 0.0)
    return (ingreso_cat / total) * 100


def productos_baja_rotacion(lista_ventas: list[dict], cantidad_maxima: int = 2) -> list[str]:
    """Devuelve nombres de productos con cantidad vendida <= cantidad_maxima."""
    return [v["producto"] for v in lista_ventas if v["cantidad"] <= cantidad_maxima]

# Ejemplos de uso
print("Ingresos totales:", calcular_ingresos_totales(ventas))
print("Producto mÃ¡s vendido:", producto_mas_vendido(ventas))
print("Ticket promedio:", calcular_ticket_promedio(ventas))
print("Ventas por categorÃ­a:", ventas_por_categoria(ventas))
print("CategorÃ­a top:", categoria_top(ventas))
print("Productos premium:", productos_premium(ventas))
print("% ElectrÃ³nica:", calcular_porcentaje_categoria(ventas, "ElectrÃ³nica"))
print("Productos de baja rotaciÃ³n:", productos_baja_rotacion(ventas))
````
